```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("S1AIW_S2AL2A_DEM_IDX_SOCS_SG_300m.csv")
df2 = pd.read_csv("land1_11_inven_300m.csv")
df3 = pd.read_csv("agb_only_geom.csv")
```
